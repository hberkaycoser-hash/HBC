```python
import requests
import os
from google.colab import drive
import time
import numpy as np

# Google Drive'a bağlan
drive.mount('/content/drive')

# Kaydedilecek ana klasör (Google Drive içindeki konum)
drive_folder = "/content/drive/MyDrive/colab_notebooks/Berkay/Safe_Unsafe_Street/Elifsu"

# API Key
API_KEY = "AIzaSyBoHh-k5WY11bZRaOPmtIB8tPvd1o2MO58"

# Sokak başlangıç ve bitiş koordinatları
locations = {
    "Üsküdar - Kuzguncuk Mahallesi":{
        "Icadiye_Street": [(41.036359, 29.029453), (41.030351, 29.035331)],
        "Perihan_Abla_Street": [(41.035953, 29.029992), (41.036088, 29.030957)],
        "Uryanizade_Street": [(41.036689, 29.030123), (41.035491, 29.030361)],
        "Simitci_Tahir_Street": [(41.033792, 29.031414), (41.032911, 29.032586)],
        "Bostan_Street": [(41.034582, 29.030860), (41.035456, 29.031085)],
  },

}

# İki koordinat arasında 12 eşit nokta oluştur
def interpolate_coordinates(start, end, num_points=12):
    lat_points = np.linspace(start[0], end[0], num_points)
    lon_points = np.linspace(start[1], end[1], num_points)
    return list(zip(lat_points, lon_points))

# Google Street View'dan fotoğraf çekme fonksiyonu
def download_street_view_image(lat, lon, heading, image_index, street_name, neighborhood):
    url = f"https://maps.googleapis.com/maps/api/streetview?size=800x600&location={lat},{lon}&heading={heading}&key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        # Mahalle ve sokak için klasör oluştur
        neighborhood_folder = os.path.join(drive_folder, neighborhood)
        street_folder = os.path.join(neighborhood_folder, street_name)
        os.makedirs(street_folder, exist_ok=True)

        # Fotoğrafı kaydet
        image_path = os.path.join(street_folder, f"{street_name}_{image_index}_heading{heading}.jpg")
        with open(image_path, "wb") as file:
            file.write(response.content)
        print(f"✅ {street_name} ({image_index}, {heading}°) kaydedildi: {image_path}")
    else:
        print(f"❌ {street_name} ({image_index}, {heading}°) için fotoğraf alınamadı! (Kod: {response.status_code})")

# Sokaklar üzerinde işlem yapma
for neighborhood, streets in locations.items():
    for street_name, (start_coord, end_coord) in streets.items():
        interpolated_coords = interpolate_coordinates(start_coord, end_coord, num_points=12)

        for i, (lat, lon) in enumerate(interpolated_coords):
            for heading in [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360]:  # Her noktadan 13 farklı açı
                download_street_view_image(lat, lon, heading, i, street_name, neighborhood)
                time.sleep(1)  # API sınırlarını aşmamak için bekleme süresi

print("🚀 Tüm görseller indirildi ve Drive'a kaydedildi!")
```
